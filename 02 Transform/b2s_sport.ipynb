```python
from pyspark.sql import Row
from pyspark.sql.types import StructType, StructField, LongType, StringType, DoubleType, BooleanType

spark.sql("DROP TABLE IF EXISTS strava.silver.sport")

sport_schema = StructType([
    StructField("id", StringType(), True),
    StructField("name", StringType(), True),
    StructField("type", StringType(), True),
     StructField("description", StringType(), True)
])

sport_rows = [
    Row(id=999, name="Dummy Sport", type="Test", description="This is a dummy row"),
    Row(id=1000, name="Sample Sport", type="Example", description="Sample description")
]

sport_df = spark.createDataFrame(sport_rows, schema=sport_schema)
sport_df.write.format("delta").mode("overwrite").saveAsTable("strava.silver.sport")
display(sport_df)
```
