```python
from pyspark.sql import Row
from pyspark.sql.types import StructType, StructField, LongType, StringType, DoubleType, BooleanType

spark.sql("DROP TABLE IF EXISTS strava.silver.gear")

gear_schema = StructType([
    StructField("id", StringType(), True),
    StructField("name", StringType(), True),
    StructField("distance_km", DoubleType(), True),
    StructField("moving_time_s", LongType(), True),
    StructField("elevation_gain_m", DoubleType(), True),
    StructField("device_watts", BooleanType(), True),
    StructField("description", StringType(), True),
    StructField("weight", DoubleType(), True),
    StructField("created_at", StringType(), True),
    StructField("updated_at", StringType(), True)
])

gear_data = [Row(
    id="g12345",
    name="Placeholder Bike",
    distance_km=1200.5,
    moving_time_s=36000,
    elevation_gain_m=15000.0,
    device_watts=True,
    description="Carbon frame road bike",
    weight=7.8,
    created_at="2025-01-01T10:00:00Z",
    updated_at="2025-12-31T15:00:00Z"
)

gear_df = spark.createDataFrame(gear_data, schema=gear_schema)
gear_df.write.format("delta").mode("overwrite").saveAsTable("strava.silver.gear")
display(gear_df)
```
