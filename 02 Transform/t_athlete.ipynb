Placeholder dimension table for athlete details, which will be extracted from source later on with -- get DetailedAthlete -- postman command

```python
from pyspark.sql import Row
from pyspark.sql.types import StructType, StructField, LongType, StringType

spark.sql("DROP TABLE IF EXISTS strava.silver.athlete")

schema = StructType([
    StructField("id", LongType(), True),
    StructField("firstname", StringType(), True),
    StructField("lastname", StringType(), True),
    StructField("city", StringType(), True),
    StructField("country", StringType(), True),
    StructField("sex", StringType(), True),
    StructField("premium", StringType(), True),
    StructField("created_at", StringType(), True),
    StructField("updated_at", StringType(), True)
])

athlete_data = [Row(
    id=142655309,
    firstname="Amadej",
    lastname=None,
    city=None,
    country=None,
    sex=None,
    premium=None,
    created_at=None,
    updated_at=None
)]

athlete_df = spark.createDataFrame(athlete_data, schema=schema)
athlete_df.write.format("delta").mode("overwrite").saveAsTable("strava.silver.athlete")
display(athlete_df)
```
