```sql
-- Statement 1: Declare a variable
DECLARE VARIABLE row_limit = 10;

-- Statement 2: Create a temporary view using the variable
CREATE OR REPLACE TEMPORARY VIEW filtered_strava AS
SELECT
  *
FROM
  silver.strava
LIMIT row_limit;

-- Statement 3: Query the temporary view
SELECT
  *
FROM
  filtered_strava;

--1 find highest kudos count activity
--2 what is average speed by sport type
--3 how many activities are marked as commute, and how many as race? etc
```
